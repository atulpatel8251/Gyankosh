```python
def mcq_to_markdown(mcq_dict):
    markdown = ""
    for key, value in mcq_dict.items():
        if key.isdigit():
            markdown += f"Question {key}: {value}\n"
        elif key.startswith('image'):
            markdown += f"![Image {key[-1]}](path/to/{value})\n\n"
    return markdown
```

```python
def display_mcq(mcq_dict):
    for key, value in mcq_dict.items():
        if key.isdigit():
            st.markdown(f"**Question {key}:** {value}")
        elif key.startswith('image'):
            st.image(f"path/to/{value}", use_column_width=True)
```

```python
display_mcq(data['6 Marks Questions'])
```

```python
import docx
import json

def extract_questions_from_docx(file_path):
    doc = docx.Document(file_path)
    
    topic = ""
    questions = []

    current_question_number = None
    current_question = None
    current_options = {}

    for paragraph in doc.paragraphs:
        text = paragraph.text.strip('\n')
        print(text)
        print('hello')
        if text:
            if text.startswith("Topic :"):
                topic = text.split(":")[1].strip()
            elif text.startswith("1) "):  # Assuming question number starts with ")"
                if current_question_number is not None:
                    current_question["options"] = current_options
                    questions.append(current_question)
                
                current_question_number = text.split(") ")[0].strip()
                current_question = {"question": text.split(") ")[1]}
                current_options = {}
            elif text.startswith("("):  # Assuming options start with "("
                option_letter = text.split(") ")[0].strip()
                option_text = text.split(") ")[1]
                current_options[option_letter] = option_text

    if current_question_number is not None:
        current_question["options"] = current_options
        questions.append(current_question)

    return {"topic": topic, "questions": questions}



def main():
    file_path = "class_10_maths.docx"  # Update the file path
    data = extract_questions_from_docx(file_path)
    json_data = json.dumps(data, indent=4, ensure_ascii=False)
    print(json_data)

#     with open("output.json", "w", encoding="utf-8") as f:
#         f.write(json_data)

if __name__ == "__main__":
    main()
```
